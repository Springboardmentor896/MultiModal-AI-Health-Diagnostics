```python
import pandas as pd
from model_parameter_interpreter import classify_parameter

df = pd.read_csv("model1_ground_truth.csv")

correct = 0

for _, row in df.iterrows():
    pred = classify_parameter(float(row["Value"]), row["Parameter"])
    if pred == row["True_label"]:
        correct += 1

accuracy = (correct / len(df)) * 100
print("Model-1 Accuracy:", accuracy)
```
