```python
import pandas as pd

def validate_row(row):
    numeric = [
        "Age","Hemoglobin","Glucose","Cholesterol",
        "White Blood Cells","Red Blood Cells","Platelets"
    ]

    for col in numeric:
        if pd.isna(row[col]):
            return False
        try:
            float(row[col])
        except:
            return False

    gender = str(row["Gender"]).lower().strip()
    if gender not in ["male","female"]:
        return False

    return True
```
