```python
df = load_dataset("../Milestone-1/Reports")

for i, row in df.iterrows():
    if not validate_row(row):
        continue

    # Model-1
    status = interpret_blood_report(row)

    # Model-2
    patterns = detect_patterns(row)
    base_risk = calculate_total_risk(row)

    # Model-3
    final_risk = adjust_by_context(row, base_risk)
    risk_level = get_severity(final_risk)

    # Milestone-3
    summary = synthesize_findings(status, patterns, final_risk, risk_level)
    recs = generate_recommendations(summary, row["Age"], row["Gender"])

    print("\n" + "-"*70)
    print(f"PATIENT {i+1} | Age: {row['Age']} | Gender: {row['Gender']}")
    print("-"*70)

    print("\nSYNTHESIZED SUMMARY")
    for f in summary["key_findings"]:
        print("•", f)

    print(f"\nOverall Risk: {summary['overall_risk']}")

    print("\nRECOMMENDATIONS")
    for r in recs:
        print("•", r)

    print("-"*70)
```
