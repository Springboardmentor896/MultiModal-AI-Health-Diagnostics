```python
import pandas as pd

df = pd.read_csv("blood_dataset.csv")
df.head()

def classify_value(v):
    if v < 0.33:
        return "Low"
    elif v > 0.66:
        return "High"
    else:
        return "Normal"


parameters = [
    "Hemoglobin",
    "Glucose",
    "Cholesterol",
    "White Blood Cells",
    "Red Blood Cells"
]

rows = []

for _, row in df.iterrows():
    for p in parameters:
        rows.append({
            "Parameter": p,
            "Value": row[p],
            "True_Label": classify_value(row[p])
        })

labeled_df = pd.DataFrame(rows)
labeled_df.head()

from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(
    labeled_df, test_size=0.2, random_state=42, shuffle=True
)

train_df.to_csv("train_validation.csv", index=False)
test_df.to_csv("test_validation.csv", index=False)

print("Train size:", len(train_df))
print("Test size:", len(test_df))
```
